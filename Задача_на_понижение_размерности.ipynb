Понижение размерности с помощью метода TSNE и подбираем гипепараметр с помощью метода локтя

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.pyplot as pp
from sklearn.manifold import TSNE
import numpy as np
from sklearn.cluster import KMeans

dt = pd.read_csv('/content/data.csv')
X = dt.iloc[:1000]

from sklearn.manifold import TSNE
transformer = TSNE(n_components=2)
X2= transformer.fit_transform(X)
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i, init='k-means++',random_state=42)
    kmeans.fit(X2)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('Метод локтя')
plt.xlabel('Количество кластеров')
plt.ylabel('Ошибки')
plt.show()
```

Вывод кластеров данных

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(X2)
labels = kmeans.labels_

pp.scatter(X2[:,0],X2[:,1],c=labels)
```
